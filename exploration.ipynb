# Exploration données HPO

```python
import re

cursor = connMedicalData.cursor()
cursor.execute("""
    CREATE TABLE IF NOT EXISTS hpo_terms (
        hpo_id TEXT PRIMARY KEY,
        name TEXT,
        synonyms TEXT
    )
""")

with open("assets/HPO/hpo.obo", "r", encoding="utf-8") as f:
    content = f.read().split("[Term]")
    for term in content[1:]:
        hpo_id_match = re.search(r"^id: (HP:\d+)", term, re.M)
        name_match = re.search(r"^name: (.*)", term, re.M)

        if hpo_id_match and name_match:
            hpo_id = hpo_id_match.group(1)
            name = name_match.group(1).lower()
            synonyms = "|".join(re.findall(r'synonym: "(.*)"', term)).lower()
            cursor.execute(
                "INSERT OR REPLACE INTO hpo_terms VALUES (?, ?, ?)",
                (hpo_id, name, synonyms),
            )

connMedicalData.commit()
```

```python
#Select all synonyms of a secondary effect. Retrieve the secondary effect ID
connMedicalData.execute("SELECT * FROM hpo_terms WHERE name='fever'").fetchall()
```

On aperçoit que les données OMIM sont organisées en sections, chaque section correspondant à une maladie. Chaque section commence par un identifiant de maladie (ex: 100100) suivi de différentes propriétés (ex: Titre, Symptômes, etc.).

```python
omim_id = "108050"
details = {"id": omim_id, "symptoms": [], "title": ""}

found = False
current_field = None

with open("assets/OMIM/omim.txt", 'r', encoding='utf-8') as f:
    for raw in f:
        line = raw.rstrip("\n")

        # nv maladie
        if line.startswith('*RECORD*'):
            found = False
            current_field = None
            continue

        # maladie est celle recherchée
        if not found and line.strip() == omim_id:
            found = True
            continue

        if not found:
            continue

        if line[:7] == '*FIELD*':
            if line[8:10] == 'TI':
                current_field = 'TI'
                continue
            elif line[8:10] == 'CS':
                current_field = 'CS'
                continue
            else:
                current_field = None
                continue

        if current_field == 'TI':
            details["title"] = line.strip()
        elif current_field == 'CS':
            if line.strip():
                details["symptoms"].append(line.strip())

# result
details
```
